```python
from preprocess import create_tag_dict
import numpy as np
```

```python
tagset=set()
xls_pathname = '../CC_FILES_TAGLIST/TW Case List.xlsx'
tag_dict = create_tag_dict(xls_pathname,v=False)
index=0
for key in tag_dict.keys():
    tagdoc=" ".join(str(x) for x in tag_dict[key])
    tagset |= set(tag_dict[key])
    print("index:{} doc:{} content:{}".format(index,'tagdoc_'+str(key),tagdoc))
    index += 1
```

```python
print('tagset:')
for tag in sorted(tagset):
    print (tag)
```

```python
doc_array=[]
tag_array=[]
for key in tag_dict.keys():
    padded_doc=[]
    tag_array.append(tag_dict[key])
    for tag in sorted(tagset):
        if tag in tag_dict[key]:
            padded_doc.append(1)
        else:
            padded_doc.append(0)
    #print (padded_doc)
    doc_array.append(padded_doc)
    
X=np.array(doc_array)

#print(X)
         
           
```

```python
from sklearn.cluster import KMeans
n_clusters=5
kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
print("labels:",kmeans.labels_)
centers=kmeans.cluster_centers_
print("clusters:",kmeans.cluster_centers_)
```

```python

y=[]

for c in range(n_clusters):
    s=[]
    for d in range(len(X)):
        
        p = kmeans.predict([doc_array[d]])
        
        if p == [c]:
            
            print ("\nCluster:{} Doc Index:{}".format(c,d))
            print ("tag_content:",tag_array[d])
            print ("tag_vector[{}]:{}".format(d,doc_array[d]))
            #print ("X[{}]:{}".format(d,X[d]))
            print ("prediction:", p)
            s.append(set(tag_array[d]))
   
    y.append(set.intersection(*s))
        
for c,yc in enumerate(y):    
    print ("\nSet Intersection[{}]:{}".format(c,yc))
```
