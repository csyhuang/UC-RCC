### Create polarity files for use by CNN_Train and CNN_evaluate
### Adapted  by Frank J. Greco for analysing Atlas transcript data

```python
import json
import re
import os


def create_polarity_files(tag):
    data_dir = os.path.abspath(os.path.join(os.path.curdir, "Data_"+tag))
    if not os.path.exists(data_dir):
        os.makedirs(data_dir)

    print("Writing to {}\n".format(data_dir))

    tf=open('text_dict3.json','r')

    tpr=open(data_dir+'/train_pos.txt','wb')
    tnr=open(data_dir+'/train_neg.txt','wb')

    tpe=open(data_dir+'/test_pos.txt','wb')
    tne=open(data_dir+'/test_neg.txt','wb')

    text = json.load(tf)

    count = 0

    POS_Test = 0
    NEG_Test = 0

    POS_Train = 0
    NEG_Train = 0

    POS_flag=1
    NEG_flag=1

    for item in text.keys():

        count +=1
        line=re.sub(r'\W+', ' ', text[item]['text']) #strip whitespace
        print ("count:", count ,"len", len(text[item]['text']), len(line))

        if tag in text[item]['tags']:
            print ('POS')

            if POS_flag == 1:
                tpr.write(line+'\n')
                POS_flag=0
                POS_Train += 1
            else:
                tpe.write(line + '\n')
                POS_flag=1
                POS_Test += 1
        else:
            print ('NEG')
            if NEG_flag == 1:
                tnr.write(line+'\n')
                NEG_flag=0
                NEG_Train += 1
            else:
                tne.write(line + '\n')
                NEG_flag=1
                NEG_Test += 1

    tpr.close()
    tnr.close()
    tpe.close()
    tne.close()

    print ('POS_Train', POS_Train, 'NEG_Train', NEG_Train)
    print ('POS_Test', POS_Test, 'NEG_Test', NEG_Test)
```

```python
create_polarity_files('TW04')
```
